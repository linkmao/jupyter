### Entorno de ejecución para Javascript
Por defecto jupyter tien un entorno de ejecución para Python, (kernel) mientras que no lo trae para javascipt, es por eso que se le debe instalar un pequeño paquete para que jupyter tenga el entorno de ejecucipon para JS `ijavascript`
la instalación de `ijavascript` es sencilla
```
npm install -g ijavascript
```
**Nota importante:** Jupyter y ijavascript para que tenga el entorno de ejecución de JS solo me funciona si tiene un node con version inferior a 20, pues le instalé el último y a la fecha no era compatible.

Esto último me causó un problema, así que instalé el programam **node version manager** para tener varias versiones de node y poder cambiar entre ellas. La documentación está en la siguiente sección
  

## Node version manager nvm
Es un pequeño programa que permite gestionar diferentes versiones de node, permitiendo instalarlas, desistalarlas y cambiar la version de ser necesario

### Instalación
Su paquete extrañamente no está en npm, pero se puede conseguir de este [repositorio](https://github.com/coreybutler/nvm-windows/releases) en git hub.

Se descarga el archivo **nvm-setup.zip**, se descomprime y se instala.

### ver versiones de node intaladas
```
nvm list
```

### cambiar a una version de node
```
nvm use <version>
```
### Instalar una version de node
```
nvm install <version>
```
### Desinstalar una version de node
```
nvm uninstall <version>
```

## PosgretSql
PsgretSql es una base de datos open source que puede ser instalada en el computador para desarrollar con el. En este notebook se explica como es la instalación y el uso de unos comandos básicos de **PosgreSql**

### Instalación
Se debe instalar desde la página web de [postgresql.org](https://www.postgresql.org/) eligiendo por supuesto el sistema operativo necesario.

- Se da siguiente a todo
- Se debe elegir password para la base de datos (en mi caso será **desarrollo**)
- pide un puerto para escuchar se deja por defecto el puerto **5432**
- Al terminar de instalar solicita ejectura Stack Builder lo permitimos
- Se recomienda de la opcion que ofrece Stack Builder instalar todas las opciones de Database Divres
- Esto hace que se instale 4 veces los drivers así que paciencia

### Como ejecutar
- Usando *SQL Shell*
    las opciones por defectos son Server:localhost Database:postgresql Port:5432 Username: postgres Contraseña:desarrollo

- Desde el cmd usando.
Algo que noté es que si no se especifica o usuario o base de datos usa algunas por defecto, así que se recomienda usar ambos parámetros
```
psql -U usuario -d dataBase
```
**Importante:** el usuario principal es postgres es el super usuario y con el se pueden crear otros usuarios que podran acceeder por defecto a todas las bases de datos

### Comando básicos
#### Crear una base de datos
```
create database [name];
```
#### Listar base de datos
```
\l
```
#### Conectarse a la base de datos para usarla
```
\c [name]
```

#### Cambiar el nombre a una base de datos
```

```

#### Crear tablas en la base de datos
```

   name text,
   id int,

```
#### Listar las tablas de la base de datos
```
\dt
```
##### insertar datos a una tabla
```

'Colombia', 1 );
```

#### Ver los datos en la tabla
```
select * from equipos;
```
#### Crear un usuario
```
CREATE USER myuser WITH PASSWORD 'secret_passwd';
```

#### Ver todos los usuarios
```
\du

## MongoDb

Instalo mongo server, descargandola desde la web
Al momento de instalar de manera completa, debo **deseleccionar** iniciar mongo como servicio (eso haria que siempre mongo se ejecute cuando se inicie el pc, y este no es el caso), pues normalmente en desarrollo inciamos el servicio con 
```
mongod
```

Al instalarse mongo , para la ejecucion de sus programas desde consola se debe llevar la ruta de isntalacion de mongo la variable path del sistema.

La dirección para acceder a la base de datos localmente corresponde a
mongodb://127.0.0.1:27017/bd-name

### Inicio del servicio
Para que mongo funcione debo iniciar el servicio con 
```
mongod
```
**Nota:** Para que pueda montarse el sevidor, se debe crear la ruta c:/data/db

### Uso de shell
```
mongo
```

### Listar base da datos
```
show dbs
```

### Usar o crear una base de datos
Si la base de datos no existe la crea pero no la lista hasta que exista por lo menos una colección (tabla)
```
use [name-db]
```

### Ver base de datos actual
```
db
```

### Borrar una base de datos
```
db.dropDatabase()
```

### Crear una colección
```
db.createCollection("name")
```

### Ver las colecciones existentes en la bd
```
show collections
```

### Borrar una colección
```
db.[name-collection].drop()
```

### Agregar un documento a una coleccion (si la coleccion no existe la crea)
```
db.[name-collection].insert(json)
```

### Agregar multiples datos a una coleccion (se usa un array de json)
```
db.[name-collection].inster([json1, json2,...])
```

### Ver todos los documentos de una coleccion
```
db.[name-collections].find().[pretty()]
```
con pretty() muestra la información indentada

### Buscar un docuemnto en particular 
```
db.[name-collecions].find({prop1:"valor1", prop2:"valor2"})
```

### Buscar el primer elemento que coincida con los parametros dados
```
db.[name-collecions].findOne({prop1:"valor1", prop2:"valor2"})
```

### Filtrar el resultado de la busqueda (para que solo muestre la info que se quiera)
```
db.[name-collecions].find({prop1:"valor1", prop2:"valor2"}, {prop1:1, prop2:1, prop3:0})
```
Aquelllas propiedades que se ponen en 1 se meustran, las que se ponen en cero no se muestran.

### Mostrar una cantidad limite de documentos
```
db.[name-collectios].find({prop:"valor"}).limit(n)
```
n es la cantidad limite a mostrar

### Ver la cantidad total de documentos
```
db.[name-collectios].count()
```

### Actualizar una coleccion
```
db.[name-collection].update({prop:"data1"}, {prop:"data2"})
```
update tiene dos parametros, el primero de ello es un objeto para buscar un documento, y el segundo es el reemplazo que hará, sin embargo es importante saber que update() actualiza toda la coleccion, por lo tanto reemplaza toda la coleccion original.

### Actualizar una coleccion sin perdida de la informacion
```
db.[name-collection].update({prop:"data1"}, {$set:{prop:"data2"}})
```
con el comando $set, se logra adicionar elementos a la coleccion, o modificar un elemento ya existente en la coleccion

### Renombrar el nombre de una propiedad
```
db.[name-coellection].update({prop:"data"},{$rename:{prop:"newname"}})
```
Busca la coleccion y cambia el nomre de la propiedad que se especifica con el comando $rename

### Borrar un documento
```
db.[name-collections].remove({prop:'valor'})
```
Borra todo el documento
