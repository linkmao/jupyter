## Motor HTML
En este apartado estaré documentado la manera de usar motores HTML para el renderizado de elementos fianles de usuario como botones, cajones entre otros. La dependiencia que se describrá acá y con la cual se hizo el diseño por ejemplo de la polla 2022, es `expess-handlebars`

## Carpeta views
En esta carpeta se encuentra en su mayoria el HTML que se rendirizará sin embargo hay un par de conceptos que se debe saber. 

Cada uno de los archivos en lacarpeta view son archivos con la extensión `.hbs`, cada uno representa una vista, por ejemplo index, games, puntajes etc, es decir son cada vista principal. Particularmente la primera vista de la app se debe escribir en `index.hbs` dentro de la carpeta `view`.

Para acceder a la vista que se reqiera se realiza con peticiones get, usando el objeto res propiedad render, es decir en realidad mostrar en pantalla una vista es renderizarla. 

normalmente las peticiones get que muestran o renderizan pantallas generales se escriben en un archivo llamado `index` en la carpeta ruta, es decir es el achivo index de las rutas mas generales (login, about, etc) aunque en realidad depende del proyeto y de la arquitectura del enrutado el que define en cual porcion de codigo backend se hace esta peticion:

**Sintaxys general de renderizado de vista**
```
route.get('ruta',(req,res)=>{res.render('vista en view',{dataArray1, dataArray 2})})
```
ruta: Es la ruta en el navegador, este dispara la peticion get para renderizar la vista que se requiera
vista en view: Es el nombre del archivo .hbs en la carpeta view que se renderizará
dataArrayN: Corresponde a un **array** con datos que se llevan a la vista para su uso (mostrar usuarios, imagenes, etc) esto es fundamental pues esto es lo que permite el dinamismo de lo que se renderiza en pantalla.

Cada dataArray recuerdese que es un array estas son las opciones que he identificado
dataArray = [{un objeto json}]

**Ejemplo real**
En el archivo index.js de la carpeta route
```
route.get('/',(res,req)=>{res.render('index',{dataArray})})
```
Este renderiza el archivo `index.hbs` de la carpeta view y lleva el array dataArray para su procesamiento.

Normalmente cada view en realidad son estructuras de div, con filas y columnas, donde cada una de ellas internamente maneja o renderiza otra estructura llamada `partials`.


## Carpeta view/layout
El el concepto de layout corresponde al de una unica plantilla que se replica en todas las vistas, esto permite una cierta regularidad en las páginas web, además la ESTRUCTURA GENERAL del documento HTML va en el layout de tal manera que en las difernetes vistas no es necesario colocar de nuevo la esturctura general de HTML

el nombre del archivo dentro de la carpate layout debe ser `main.hbs`
El código que normalmente va en este archivo seria el siguiente

```
<!DOCTYPE html>
<html lang="en">

<head>
  <meta charset="UTF-8">
  <meta http-equiv="X-UA-Compatible" content="IE=edge">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>

  <!-- CSS only -->
  {{!--
  <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.2.0-beta1/dist/css/bootstrap.min.css" rel="stylesheet"
    integrity="sha384-0evHe/X+R7YkIZDRvuzKMRqM+OrBnVFBL6DOitfPri4tjfHxaWutUpFmBp4vmVor" crossorigin="anonymous"> --}}
  <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.1.3/dist/js/bootstrap.bundle.min.js"
    integrity="sha384-ka7Sk0Gln4gmtz2MlQnikT1wXgYsOg+OMhuP+IlRH9sENBO0LRn5q+8nbTov4+1p"
    crossorigin="anonymous"></script>

  <link rel="stylesheet" href="../../css/styles.css">
</head>


<body>
  <div class="container-fluid">
    {{>mensajes}}
    {{>navigation}}
    {{!-- Este body es constantemente actualizado por el llamado de cada view --}}
    {{{body}}}  
  </div>

</body>

</html>

```

El punto mas importante acá es el que está en body.. la sentencioa {{{body}}} implica que cada llamado de una vista esta se colocá allí dentro eso significa que la vista queda inmersa dentro de toda la estructura HTML que estpa en este código.

## Carpeta view/partials
Los partials son estructuras de codigo html, (quizas mas grandes que las views) que pueden usar la data que se le ha pasado a la view padre. Estos archivos son igualmente de extensión `.hbs`

A los formualrios se le asocia la ruta backend para el post o put

```
<form action="/auth/signin" method="POST">
.
.
 <div class="col text-center"><button type="submit" class="btn btn-primary mb-1 container">Ingresa</button></div>
```

Para resaltar esto pone a funcionar la ruta /auth/singin cuando el boton asociado se le de click


A los botones (difernetes de los tipo submit asociado a formualrios ) se les asigna la ruta backend para los get 
```
<div class="col text-center"><a class="btn btn-primary mb-1 container" href="/signup">Registro</a></div>
```
Para resaltar acá es la ruta signup
