# Instalacion de Clawdbot

##  Generalidades

Para la instalcion de clawdbot, inicialmente se recomienda que sea en linux, por eso este tutorial lo hace usando ese sistema operativo, sin embargo al momento de hacer estos apuntes en la documenataci√≥n se afirma que ya se puede en windows

Si estoy en windos y se quiere usar linux se puede usar wsl

Se debe intalar wsl

verificar intalacion 
wsl --version

Instalar ubuntu con wsl
wsl --install
despues de instalar se debe reiniciar

ejecutar ubunu
desde window con el buscador poner ubuntu

Ya en el shell de ububtu realizar

1. Actualizacion del SO
   
`$ sudo apt update && sudo apt upgrade -y`

2. Instalacion de nodejs (o verificacion de su intalacion con node --version)

`curl -fsSL https://deb.nodesource.com/setup_22.x | sudo -E bash -`

`sudo apt-get install -y nodejs`

3. Instalacion de Clawdbot

`curl -fsSL https://molt.bot/install.sh | bash`
