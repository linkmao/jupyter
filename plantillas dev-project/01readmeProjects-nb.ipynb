# Admon-finance

## Descripción
Desarrollo de una aplicación web usando React, el cual tiene como objetivo llevar el registro de los ingresos y los egresos de una persona, tiene como valor agregado el uso de la libreria `victory` el cual permite la creación de gráficos estadisticos como diagramas de torta, de barras entre otros.


## Tecnologías usadas
- Vite: Generador de proyectos, en este caso de react
- React
- Tailwind
- Victory: Libreria para creación de gráficos tipo diagramas de torta


**Video base :**[Video](https://www.youtube.com/watch?v=zN6xXtHww54&t=1874s)

**Repositorio:**  [Repo](https://github.com/linkmao/portafolio)

**Deploy** [Sin despliegue](*)

## Como correr
```
$ npm run dev
```

## Descripcion detallada
La aplicación que se desarrolla basicamente es un formulario que recibe dos elementos, el concepto y el valor, si el valor es positivo se considera ingreso, si es negativo se considera egreso. La información se guarda en un objeto Json, el cual persiste pues se hace uso del `localStorage` del navegador para que aunque se refreque la página los datos perduren.
Además cada dato que se registra (ingreso o egreso) se va mostrando en unas pequeñas tarjetas que pueden ser eliminadas afectando la información guardada en el JSON.

Finalmente el otro elemento importante que se usó en este proyecto fue la libreria `victory` la cual permite la creación de diversos gráficos de tipo estadistico, lo que le a un valor agregado a las aplicaciones que se desarrollan y que requiera de este elemento.

Este proyecto es importante estudiarlo muy bien pues hay conceptos de REact bien importantes asiciado al estado de los componentes, especialmente el uso de `createContext, useContext, useEffect, useReducer`.

## Conceptos usados
-   uso del contexto de React para un acceso global de los datos
-   uso de reduce de React, el cual permite la creación de funciones para la manipulación de Implementaciones y mejoras futuras

## Mis detalles técnicos

### Instalación de tailwind
para instalar Tailwind se recomienda ir a la web, dar en GER STARTED -- Framework gudes, en el cual se debe elgir el framework en el que se hata trabajado en nuesto caso vite
se encontrará un codigo así

nota: las dos primeras lineas mostradas en la guia ya no son necesarias pues ya se habia instalado el proyecto
```
$ npm create vite@latest my-project -- --template react
$ cd my-project
$ npm install -D tailwindcss postcss autoprefixer
$ npx tailwindcss init -p
```
hay otras indicaciones pero lo mas recoenadble es que se visite la página
https://tailwindcss.com/docs/guides/vite

### Instalacion de Victory
Como novedades en este proyecto es que se usa una libreria para la generación de gráficos de datos, se googlea victory react y nos lleva a  la pagina 
https://formidable.com/open-source/victory/

y alli se entra a get started, donde basicamente indica que para instalar victory se usa el comando.
```
$ npm install victory
```
